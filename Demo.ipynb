### Instructions

For 4 base datasets Mutag, Cyclicity, Motif and Shape, this demo notebook can be used to train GNNInterpreters.
* Set explain_all_classes True to train a GNNInterpreter for all classes in the dataset.
* If you want to train for specific classes set the corresponding explain_<class> parameters True and set explain_all_classes False.
* Generated explanation graphs will be saved in <Dataset_name>DatasetsPlots folder.
* Training plots will be shown and saved at training_plots folder.
* Seeds must be provided in a list.
* To get detailed metrics run desired datasets with the default paths, all classes and at least 2 seeds. 
* A new GNN model can be tested by substracting the pre_trained_model_checkpoint_path.
* Logs will be overwritten with each run, use a different path if you want to avoid overriding it.

### Calculating Accuracy

```python
# Accuracy
xgnn_class_probs_log_path="logs/class_probs_xgnn.txt"
xgnn_train_times_log_path="logs/train_times_xgnn.txt"
accuracy_paths = [mutag_class_probs_log_path, cyclicity_class_probs_log_path, motif_class_probs_log_path, shape_class_probs_log_path]
    
for path in accuracy_paths:
    print(path[5:-4].upper())
    if 'xgnn' in path:
        log_reader.calculate_average_acc_xgnn(path)
    else:
        log_reader.calculate_average_acc(path)
    
    # Training time
    training_time_paths = [mutag_train_times_log_path, cyclicity_train_times_log_path, motif_train_times_log_path, shape_train_times_log_path]

    for _, file_path in enumerate(training_time_paths):
        total_train_time, average_train_time = log_reader.calculate_average_train_time(file_path)
        if average_train_time is not None:
            datasetName = file_path.split("_")[-1].split(".")[0]
            print(f"Average Train Time {datasetName}: ", average_train_time)
            print(f"Total Train Time{datasetName}: {total_train_time}")
```
